```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
from ipywidgets import interact, FloatSlider, HBox, VBox, interactive
import ipywidgets as widgets
```

```python
wm = FloatSlider(min=-2, max=2, step=0.1)
wb = FloatSlider(min=-2, max=2, step=0.1)
```

```python
%matplotlib ipympl
x = np.linspace(-10,10,1000)
fig, ax = plt.subplots(figsize=(4,4), dpi=136)
line = ax.plot(x,x)[0]
ax.grid()
ax.set(xlim=(-5,5), ylim=(-5,5))
```

```python
def update_line(m=1, b=0):
    line.set_ydata(m*x+b)
#    fig.canvas.draw_idle()
    fig.canvas.draw()

#interact(update_line, m=wm, b=wb);
```

```python
for i in np.arange(-2,2.001,0.2):
    update_line(i,0)
    time.sleep(0.1)
```
