```python
from collections import deque
import numpy as np
from numpy import sin, pi, zeros
import matplotlib.pyplot as plt
import time
```

```python
def g(x, T=20):
    f = 1/T
    y=sin(2*pi*f*x)
    return(y)
```

```python
%matplotlib ipympl
fig, (axRX, axTX) = plt.subplots(1,2,figsize=(14,6))
x = 0
y = g(x)
dataRX = deque([(x, y)], maxlen=20)
lineRX, = axRX.plot(*zip(*dataRX), "b.")

dataTX = deque([(x, y)], maxlen=20)
lineTX, = axTX.plot(*zip(*dataRX), "r.")
```

```python
def update_data_test():
    global x
#    global dataRX
    x += 1
    y = g(x)

    dataRX.append((x, y))
    lineRX.set_data(*zip(*dataRX))
    axRX.relim()
    axRX.autoscale_view()

    dataTX.append((x, y))
    lineTX.set_data(*zip(*dataTX))
    axTX.relim()
    axTX.autoscale_view()

    fig.canvas.draw()

    #    time.sleep(0.08)
```

```python
t1 = time.time()
N = 100
for i in range(N): update_data_test()
t2 = time.time()
print(f"Time elapsed for {N} iterations: {t2 - t1:4.2f} s")
```
