```python
# Min velocity: 1.5 m/min
# Max velocity: 3.0 m/min
vmin = 1.5/60 # m/s
vmax = 3.0/60 # m/s
print(f"{vmin=:.3f} m/s")
print(f"{vmax=:.3f} m/s")
```

```python
d = 9e-2 # 9cm pot diameter
t = d/vmax # how long it taes to pass a pot
print(f"{t=:.3f} s")
```

```python
# sampling freqency: 
fs = 10 # 1/s
ts = 1/fs # sampling interval
ds = vmax*ts # max distance between samples
print(f"{ds=:.3f} m")
```

```python
import keyboard
```

```python
x = y = 0
xold = yold = -1
```

```python
while True:
    if keyboard.is_pressed("q"): 
        yold = y 
        y+=1
    if keyboard.is_pressed("a"): 
        yold = y
        y-=1
    if keyboard.is_pressed("k"): 
        xold = x
        x+=1
    if keyboard.is_pressed("j"): 
        xold = x
        x-=1
    if (yold != y) | (xold != x):
        xold = x ; yold = y       
        print(f"{x=:6d}   {y=:6d}")
    sleep(0.1)
```
